```python
df= pd.read_csv("smartphones_Cleaned_data.csv")
```

```python
df['rating'] = df['rating'].fillna(df['rating'].median())
```

```python
df['Processor_core'] = df['Processor_core'].replace('', np.nan)
```

```python
df['Processor_core'] = df['Processor_core'].fillna('Not Available')
```

```python
df['Processor_speed'] = df['Processor_speed'].replace('',np.nan)
```

```python
df['Processor_speed'] = (
    df['Processor_speed']
    .str.replace('Processor', '', regex=False)
    .astype(float)
)
```

```python
df['Processor_speed'].median()
```

```python
df['Processor_speed'] = df['Processor_speed'].fillna(df['Processor_speed'].median())
```

```python
df['rom'] = df['rom'].replace('', np.nan)
```

```python
df['rom'] = (
    df['rom']
    .astype(str)
    .str.replace('[^0-9]', '', regex=True)  # keep only numbers
    .replace('', np.nan)
    .astype(float)
)
```

```python
df['rom'] = df['rom'].fillna(df['rom'].median())
```

```python
df['Fast_charging'].head()
```

```python
df['Fast_charging'] = df['Fast_charging'].replace('', np.nan)
```

```python
df['Fast_charging'] = (
    df['Fast_charging']
    .astype(str)
    .str.lower()
)
```

```python
df['Fast_charging'] = df['Fast_charging'].fillna('Not Supported')
```

```python
df['Battery_capacity'] = df['Battery_capacity'].replace('', np.nan)
```

```python
df['Battery_capacity'] = (df['Battery_capacity'].astype(str).str.replace('[^0-9]','',regex= True))
```

```python
df['Battery_capacity'] = df['Battery_capacity'].replace('',np.nan).astype(float)
```

```python
df['Battery_capacity'] = df['Battery_capacity'].fillna(
    df['Battery_capacity'].median()
)
```

```python
df['Battery_Power'] = (
    df['Battery_Power']
    .astype(str)
    .str.replace('[^0-9]', '', regex=True)  # keep only numbers
    .replace('', np.nan)
)
```

```python
df['Battery_Power'] = df['Battery_Power'].astype(float)
```

```python
df['Battery_Power'] = df['Battery_Power'].fillna(
    df['Battery_Power'].median()
)
```
